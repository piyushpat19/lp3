```python
def knapsack(values, weights, capacity):
    n = len(values)
    # Create DP table initialized to 0
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    # Fill the DP table
    for item in range(1, n + 1):
        for weight in range(1, capacity + 1):
            if weights[item - 1] <= weight:
                dp[item][weight] = max(
                    dp[item - 1][weight - weights[item - 1]] + values[item - 1],
                    dp[item - 1][weight]
                )
            else:
                dp[item][weight] = dp[item - 1][weight]

    return dp


def find_selected_items(dp, weights, values, capacity):
    i = len(values)
    w = capacity
    selected = []

    # Backtrack to find which items were chosen
    while i > 0 and w > 0:
        if dp[i][w] != dp[i - 1][w]:
            selected.append(i - 1)  # item index
            w -= weights[i - 1]
        i -= 1

    return selected[::-1]  # reverse for correct order


# Main loop
while True:
    print("\n=== 0/1 Knapsack Problem ===")
    n = int(input("Enter number of items: "))

    values = [int(i) for i in input("Enter values of items: ").split()]
    weights = [int(i) for i in input("Enter weights of items: ").split()]

    if len(values) != n or len(weights) != n:
        print("❌ Error: Number of values and weights must match number of items.")
        continue

    capacity = int(input("Enter maximum weight capacity: "))

    # Run knapsack
    dp = knapsack(values, weights, capacity)
    max_value = dp[-1][-1]
    selected_items = find_selected_items(dp, weights, values, capacity)

    print("\n✅ Results:")
    print("Maximum value that can be carried:", max_value)
    print("Selected items (0-indexed):", selected_items)

    print("\nItem details:")
    for idx in selected_items:
        print(f"  Item {idx + 1}: Value = {values[idx]}, Weight = {weights[idx]}")

    # Option to exit
    choice = input("\nDo you want to try again? (y/n): ").strip().lower()
    if choice != 'y':
        print("Exiting program. Goodbye!")
        break
 #Enter number of items: 3
#Enter values of items: 60 100 120
#Enter weights of items: 10 20 30
#Enter maximum weight: 50
#The maximum value of items that can be carried: 220
```
